```python
import seaborn as sns
import matplotlib.pyplot as plt

# Generate some random data
data = sns.load_dataset("tips")

# Create a KDE plot
sns.kdeplot(data["total_bill"], shade=True)

# Add labels and a title
plt.xlabel('Total Bill Amount')
plt.ylabel('Density')
plt.title('KDE Plot Example')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")["total_bill"]

# Create a distribution plot (histogram + KDE)
sns.distplot(data, bins=30, hist=True, kde=True, color='blue')

# Add labels and a title
plt.xlabel('Total Bill Amount')
plt.ylabel('Density/Frequency')
plt.title('Distribution Plot Example')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = sns.load_dataset("tips")

# Create a joint plot
sns.jointplot(x="total_bill", y="tip", data=data, kind="scatter")

# Add labels and a title
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.title('Joint Plot Example - Scatter')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (correlation matrix)
data = sns.load_dataset("flights")
flights_data = data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(flights_data, cmap="YlGnBu", annot=True, fmt="d")

# Add a title
plt.title("Flights Data Heatmap")

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (correlation matrix)
data = sns.load_dataset("flights")
flights_data = data.pivot_table(index='month', columns='year', values='passengers')

# Create a clustermap
sns.clustermap(flights_data, cmap="YlGnBu", annot=True, fmt="d")

# Add a title
plt.title("Flights Data Clustermap")

# Show the plot
plt.show()
```
